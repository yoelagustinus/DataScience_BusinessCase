```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import datetime
import xlrd
```

```python
workbook_sls = xlrd.open_workbook('sales_data.xls')
```

```python
worksheet_sls = workbook_sls.sheet_by_name('Sheet 1')
```

```python
data_sls = []
header_sls = []
for col_sls in range(worksheet_sls.ncols):
    cell_value = worksheet_sls.cell_value(0, col_sls)
    header_sls.append(cell_value)

for row_sls in range(1, worksheet_sls.nrows):
    row_data_sls = []
    for col_sls in range(worksheet_sls.ncols):
        cell_value = worksheet_sls.cell_value(row_sls, col_sls)
        row_data_sls.append(cell_value)
    data_sls.append(row_data_sls)
```

```python
df_sales = pd.DataFrame(data_sls, columns = header_sls)
df_sales
```

```python
df_sales.describe(include='all')
```
